```python
# convert fall_state column to interger
sensor['fall_state'] = sensor['fall_state'].astype(int)
```

```python
import plotly.express as px

fig = px.line(sensor, x=sensor.index, y=["ax", "ay", "az", "fall_state"], title='Sensor data')
fig.show()
```

```python
import plotly.graph_objs as go
from plotly.subplots import make_subplots

# Create a figure with subplots
fig = make_subplots(rows=2, cols=1)

# First subplot for 'ax', 'ay', 'az'
fig.add_trace(go.Scatter(x=sensor.index, y=sensor['ax'], name='ax'), row=1, col=1)
fig.add_trace(go.Scatter(x=sensor.index, y=sensor['ay'], name='ay'), row=1, col=1)
fig.add_trace(go.Scatter(x=sensor.index, y=sensor['az'], name='az'), row=1, col=1)

# Second subplot for 'fall_state'
fig.add_trace(go.Scatter(x=sensor.index, y=sensor['fall_state'], name='fall_state'), row=2, col=1)

# Update layout if needed
fig.update_layout(height=600, width=1000, title_text="Sensor Data")
fig.show()
```
